```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("weather_data_700.csv")
```

```python
df
```

```python
plt.scatter(df['date'], df['meantemp'], color = 'red')
```

```python
x1 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['EWMA'] = x1
df
```

```python
plt.scatter(df['date'], df['EWMA'], color = 'red')
plt.plot(df['date'], x1, color='black', label='alpha = 0.9')
plt.legend()
plt.show()
```
