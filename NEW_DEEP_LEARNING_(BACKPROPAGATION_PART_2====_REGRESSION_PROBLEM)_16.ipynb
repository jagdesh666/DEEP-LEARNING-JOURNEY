INITIALIZING THE VALUE FOR WEIGHTS AND BIASES

```python
def initialize_parameter(layer_dims):

  np.random.seed(3)
  parameters = {}
  L = len(layer_dims)

  for l in range(1, L):

    parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
    parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))

  return parameters
```

```python
initialize_parameter([2,2,1])
```

DOT PRODUCT

```python
def linear_forward(A_prev, W, b):

  Z = np.dot(W.T, A_prev) + b

  return Z
```

IMPLEMENTING FORWARD PROPAGATION

```python
def l_layer_forward(x , parameter):

  a = x
  L = len(parameter) // 2

  for l in range(1, L+1):
    a_prev = a
    Wl = parameter['W' + str(l)]
    bl = parameter['b' + str(l)]

    print("A"+str(l-1)+": ", a_prev)
    print("W"+str(l)+": ", Wl)
    print("b"+str(l)+": ", bl)
    print("--"*20)


    a = linear_forward(a_prev, Wl, bl)
    print("A"+str(l)+": ", a)
    print("**"*20)


    return a, a_prev
```

UPDATING THE WEIGHTS AND BIASES VALUES

```python
def update_parameters(parameters, y, y_hat, x):

  paramter['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat) * A1[0][0])
  paramter['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat) * A1[1][0])
  paramter['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))


  paramter['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat) * paramter['W2'][0][0] * x[0][0])
  paramter['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat) * paramter['W2'][0][0] * x[1][0])
  paramter['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat) * paramter['W2'][0][0])


  paramter['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat) * paramter['W2'][1][0] * x[0][0])
  paramter['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat) * paramter['W2'][1][0] * x[1][0])
  paramter['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat) * paramter['W2'][1][0])
```

EPOCHS IMPLEMENTATION

```python
paramter = initialize_parameter([2,2,1])

epochs = 5

for i in range(epochs):


  loss = []

  for j in range(dataset.shape[0]):

    x = dataset[["CGPA", "PROFILE_SCORE"]].values[j].reshape(2,1)
    y = dataset[["PACKAGE"]].values[j][0]

    y_hat, A1 = l_layer_forward(x, paramter)

    y_hat = y_hat[0][0]

    update_parameters(paramter, y, y_hat, x)

    loss.append(np.square(y - y_hat)**2)

  print("Epoch ------------------------------------ ", i+1, "Loss ----------------------- ", np.array(loss).mean())








```
