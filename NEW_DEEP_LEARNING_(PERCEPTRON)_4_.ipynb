IMPORTING LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

IMPORTING DATASET

```python
dataset = pd.read_csv('student_dataset.csv')


print(dataset.shape)
print(dataset.head())
```

```python
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=dataset)
```

```python
x = dataset.iloc[:,0:2]
y = dataset.iloc[:,2]
```

IMPORTING PERCEPTRON

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

TRAINING PERCEPTRON

```python
p.fit(x,y)
```

VALUES OF WEIGHT

```python
p.coef_
```

VALUE OF BIAS (b)

```python
p.intercept_
```

VISUALIZING THE DATA

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x.values, y.values, clf=p, legend=2)
```
